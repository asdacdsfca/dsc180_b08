```python
def safe_float(value):
    """Safely convert value to float"""
    try:
        return float(value) if value is not None else np.nan
    except (ValueError, TypeError):
        return np.nan

def get_fundamental_metrics(ticker):
    """Get fundamental metrics for a given ticker"""
    try:
        stock = yf.Ticker(ticker)
        info = stock.info

        metrics = {
            "Ticker": ticker,
            "P/E Ratio (Trailing)": safe_float(info.get("trailingPE")),
            "P/E Ratio (Forward)": safe_float(info.get("forwardPE")),
            "P/B Ratio": safe_float(info.get("priceToBook")),
            "Dividend Yield": safe_float(info.get("dividendYield")),
            "Free Cash Flow": safe_float(info.get("freeCashflow")),
            "Debt to Equity": safe_float(info.get("debtToEquity")),
            "Market Cap": safe_float(info.get("marketCap")),
            "Revenue": safe_float(info.get("totalRevenue")),
            "EPS (Trailing)": safe_float(info.get("trailingEps")),
            "ROE": safe_float(info.get("returnOnEquity")),
            "Enterprise Value": safe_float(info.get("enterpriseValue")),
            "Gross Margin": safe_float(info.get("grossMargins")),
            "Operating Margin": safe_float(info.get("operatingMargins")),
            "Net Income": safe_float(info.get("netIncomeToCommon")),
            "EBITDA": safe_float(info.get("ebitda"))
        }

        return pd.Series(metrics)
    except Exception as e:
        print(f"Error getting metrics for {ticker}: {str(e)}")
        return pd.Series()
```

```python
# Calculate the median for each metric as fixed criteria
criteria = df_fundamentals.drop(columns=["Ticker"]).median(numeric_only=True)

# Define conditions for each metric
conditions = [
    df_fundamentals["P/E Ratio (Trailing)"] < criteria["P/E Ratio (Trailing)"],
    df_fundamentals["P/B Ratio"] < criteria["P/B Ratio"],
    df_fundamentals["Dividend Yield"] >= criteria["Dividend Yield"],
    df_fundamentals["Debt to Equity"] < criteria["Debt to Equity"],
    df_fundamentals["Free Cash Flow"] > criteria["Free Cash Flow"],
    df_fundamentals["ROE"] >= criteria["ROE"],
    df_fundamentals["Revenue"] > criteria["Revenue"],
    df_fundamentals["Gross Margin"] >= criteria["Gross Margin"],
    df_fundamentals["Operating Margin"] >= criteria["Operating Margin"],
    df_fundamentals["EPS (Trailing)"] > criteria["EPS (Trailing)"],
    df_fundamentals["Enterprise Value"] < criteria["Enterprise Value"],
    df_fundamentals["EBITDA"] > criteria["EBITDA"]
]

# Count how many conditions each stock meets
condition_met_count = sum(conditions)

# Filter stocks that meet more than half of the criteria (at least 10 out of 13)
filtered_stocks = df_fundamentals[condition_met_count >= 10]
filtered_stocks["Criteria Met"] = condition_met_count
filtered_stocks = filtered_stocks.sort_values(by=["Criteria Met", "ROE"], ascending=[False, False])

top_100 = filtered_stocks[:100].reset_index(drop=True)
top_100
```

```python
def calculate_rsi(data, period=14):
    """Calculate Relative Strength Index"""
    delta = data.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=period).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=period).mean()
    rs = gain / loss
    return 100 - (100 / (1 + rs))

def calculate_atr(df, period=14):
    """Calculate Average True Range"""
    high_low = df['High'] - df['Low']
    high_close = np.abs(df['High'] - df['Close'].shift())
    low_close = np.abs(df['Low'] - df['Close'].shift())

    ranges = pd.concat([high_low, high_close, low_close], axis=1)
    true_range = np.max(ranges, axis=1)

    return true_range.rolling(period).mean()

def calculate_macd(data, fast=12, slow=26, signal=9):
    """Calculate MACD"""
    exp1 = data.ewm(span=fast, adjust=False).mean()
    exp2 = data.ewm(span=slow, adjust=False).mean()
    macd = exp1 - exp2
    signal_line = macd.ewm(span=signal, adjust=False).mean()
    return macd, signal_line

def calculate_signal_score(df, i, position_open, consecutive_losses):
    """Calculate a score for buy/sell signals based on multiple factors"""
    try:
        signals = []

        # 1. Trend Signals (weight: 3)
        if not pd.isna(df['EMA_5'].iloc[i]) and not pd.isna(df['SMA_10'].iloc[i]):
            signals.append({
                'signal': df['EMA_5'].iloc[i] > df['SMA_10'].iloc[i],
                'weight': 3,
                'name': 'Short Trend'
            })

        if not pd.isna(df['SMA_10'].iloc[i]) and not pd.isna(df['SMA_30'].iloc[i]):
            signals.append({
                'signal': df['SMA_10'].iloc[i] > df['SMA_30'].iloc[i],
                'weight': 3,
                'name': 'Medium Trend'
            })

        # 2. Momentum Signals (weight: 2)
        if not pd.isna(df['RSI'].iloc[i]):
            signals.append({
                'signal': 30 <= df['RSI'].iloc[i] <= 70,
                'weight': 2,
                'name': 'RSI'
            })

        if not pd.isna(df['ROC'].iloc[i]):
            signals.append({
                'signal': df['ROC'].iloc[i] > 0,
                'weight': 2,
                'name': 'ROC'
            })

        # 3. Volume Signals (weight: 2)
        if not pd.isna(df['Volume'].iloc[i]) and not pd.isna(df['Volume_MA'].iloc[i]):
            signals.append({
                'signal': df['Volume'].iloc[i] > df['Volume_MA'].iloc[i] * 1.1,
                'weight': 2,
                'name': 'Volume'
            })

        # Risk Management Adjustments
        risk_penalty = 0
        if consecutive_losses >= 2:
            risk_penalty = -0.2

        # Calculate weighted score
        total_weight = sum(signal['weight'] for signal in signals)
        if total_weight == 0:
            return 0

        weighted_sum = sum(signal['signal'] * signal['weight'] for signal in signals)
        base_score = weighted_sum / total_weight
        final_score = base_score + risk_penalty

        return max(0, min(1, final_score))

    except Exception as e:
        print(f"Error in calculate_signal_score: {str(e)}")
        return 0

def simulate_trading(ticker, start_date, end_date, strategy='A'):
    """Simulate trading with enhanced Strategy B"""
    try:
        stock = yf.Ticker(ticker)
        df = stock.history(start=start_date, end=end_date)

        if len(df) == 0:
            return None

        # Calculate indicators
        df['SMA_10'] = df['Close'].rolling(window=10).mean()
        df['SMA_30'] = df['Close'].rolling(window=30).mean()
        df['EMA_5'] = df['Close'].ewm(span=5, adjust=False).mean()
        df['RSI'] = calculate_rsi(df['Close'])
        df['Volume_MA'] = df['Volume'].rolling(window=20).mean()
        df['ROC'] = df['Close'].pct_change(periods=3)
        macd, signal = calculate_macd(df['Close'])
        df['MACD'] = macd
        df['Signal'] = signal
        df['MACD_Hist'] = macd - signal

        # Initialize trading variables
        cash = 100000
        shares = 0
        portfolio_values = []
        trades = []
        position_open = False
        last_trade_price = 0
        consecutive_losses = 0

        for i in range(1, len(df)):
            if pd.isna(df['Close'].iloc[i]) or pd.isna(df['Volume'].iloc[i]):
                continue

            current_price = df['Close'].iloc[i]

            if strategy == 'A':
                # Strategy A (Majority Vote)
                signal_score = calculate_signal_score(df, i, position_open, consecutive_losses)
                buy_threshold = 0.6
                sell_threshold = 0.4

                buy_signal = not position_open and signal_score >= buy_threshold
                sell_signal = position_open and signal_score <= sell_threshold

            else:
                # Enhanced Strategy B (MACD + RSI + Volume)
                rsi = df['RSI'].iloc[i]
                volume_surge = df['Volume'].iloc[i] > df['Volume_MA'].iloc[i] * 1.05
                macd_cross = (
                    df['MACD_Hist'].iloc[i-1] < 0 and
                    df['MACD_Hist'].iloc[i] > 0
                ) if i > 0 else False

                macd_cross_down = (
                    df['MACD_Hist'].iloc[i-1] > 0 and
                    df['MACD_Hist'].iloc[i] < 0
                ) if i > 0 else False

                # More lenient conditions for Strategy B
                buy_signal = (
                    not position_open and
                    (macd_cross or (df['MACD'].iloc[i] > df['Signal'].iloc[i])) and
                    rsi < 75 and  # Increased from 70
                    volume_surge
                )

                sell_signal = (
                    position_open and
                    (
                        macd_cross_down or
                        rsi > 75 or  # Increased from 70
                        current_price < last_trade_price * 0.97  # 3% stop loss
                    )
                )

            # Execute trades
            if buy_signal and cash > 0:
                shares_to_buy = cash // current_price
                if shares_to_buy > 0:
                    shares = shares_to_buy
                    cash -= shares * current_price
                    position_open = True
                    last_trade_price = current_price

                    trades.append({
                        'Date': df.index[i],
                        'Type': 'BUY',
                        'Price': current_price,
                        'Shares': shares,
                        'RSI': df['RSI'].iloc[i],
                        'MACD_Hist': df['MACD_Hist'].iloc[i],
                        'Volume_Ratio': df['Volume'].iloc[i] / df['Volume_MA'].iloc[i]
                    })

            elif sell_signal and shares > 0:
                cash += shares * current_price
                trade_return = (current_price - last_trade_price) / last_trade_price

                if trade_return < 0:
                    consecutive_losses += 1
                else:
                    consecutive_losses = 0

                trades.append({
                    'Date': df.index[i],
                    'Type': 'SELL',
                    'Price': current_price,
                    'Shares': shares,
                    'RSI': df['RSI'].iloc[i],
                    'MACD_Hist': df['MACD_Hist'].iloc[i],
                    'Volume_Ratio': df['Volume'].iloc[i] / df['Volume_MA'].iloc[i],
                    'Return': trade_return
                })

                shares = 0
                position_open = False

            portfolio_values.append(cash + (shares * current_price))

        # Close any open position at the end
        if shares > 0:
            cash += shares * df['Close'].iloc[-1]
            trade_return = (df['Close'].iloc[-1] - last_trade_price) / last_trade_price
            trades.append({
                'Date': df.index[-1],
                'Type': 'SELL',
                'Price': df['Close'].iloc[-1],
                'Shares': shares,
                'RSI': df['RSI'].iloc[-1],
                'MACD_Hist': df['MACD_Hist'].iloc[-1],
                'Volume_Ratio': df['Volume'].iloc[-1] / df['Volume_MA'].iloc[-1],
                'Return': trade_return
            })

        return {
            'Final_Return': (portfolio_values[-1] - 100000) / 100000,
            'Trades': trades,
            'Portfolio_Values': portfolio_values
        }

    except Exception as e:
        print(f"Error in trading simulation for {ticker}: {str(e)}")
        return None

def analyze_results(trades_df):
    """Analyze trading results with complete metrics"""
    if trades_df is None or len(trades_df) == 0:
        return {
            'Total_Trades': 0,
            'Winning_Trades': 0,
            'Losing_Trades': 0,
            'Average_Return': 0.0,
            'Best_Trade': 0.0,
            'Worst_Trade': 0.0,
            'Average_Hold_Time': 0,
            'Win_Rate': 0.0,
            'Profit_Factor': 0.0,
            'Average_Win': 0.0,
            'Average_Loss': 0.0,
            'Risk_Reward_Ratio': 0.0
        }

    try:
        buy_trades = trades_df[trades_df['Type'] == 'BUY']
        sell_trades = trades_df[trades_df['Type'] == 'SELL']

        if 'Return' not in sell_trades.columns:
            return {
                'Total_Trades': len(trades_df) // 2,
                'Winning_Trades': 0,
                'Losing_Trades': 0,
                'Average_Return': 0.0,
                'Best_Trade': 0.0,
                'Worst_Trade': 0.0,
                'Average_Hold_Time': 0,
                'Win_Rate': 0.0,
                'Profit_Factor': 0.0,
                'Average_Win': 0.0,
                'Average_Loss': 0.0,
                'Risk_Reward_Ratio': 0.0
            }

        winning_trades = sell_trades[sell_trades['Return'] > 0]
        losing_trades = sell_trades[sell_trades['Return'] <= 0]

        avg_win = winning_trades['Return'].mean() if len(winning_trades) > 0 else 0
        avg_loss = losing_trades['Return'].mean() if len(losing_trades) > 0 else 0

        total_profit = winning_trades['Return'].sum() if len(winning_trades) > 0 else 0
        total_loss = abs(losing_trades['Return'].sum()) if len(losing_trades) > 0 else 0

        profit_factor = total_profit / total_loss if total_loss != 0 else 0

        try:
            avg_hold_time = (sell_trades['Date'] - buy_trades['Date']).mean().days
        except:
            avg_hold_time = 0

        analysis = {
            'Total_Trades': len(trades_df) // 2,
            'Winning_Trades': len(winning_trades),
            'Losing_Trades': len(losing_trades),
            'Average_Return': sell_trades['Return'].mean(),
            'Best_Trade': sell_trades['Return'].max(),
            'Worst_Trade': sell_trades['Return'].min(),
            'Average_Hold_Time': avg_hold_time,
            'Win_Rate': len(winning_trades) / len(sell_trades) if len(sell_trades) > 0 else 0,
            'Profit_Factor': profit_factor,
            'Average_Win': avg_win,
            'Average_Loss': avg_loss,
            'Risk_Reward_Ratio': abs(avg_win / avg_loss) if avg_loss != 0 else 0
        }

        return analysis

    except Exception as e:
        print(f"Error in analyze_results: {str(e)}")
        return {
            'Total_Trades': 0,
            'Winning_Trades': 0,
            'Losing_Trades': 0,
            'Average_Return': 0.0,
            'Best_Trade': 0.0,
            'Worst_Trade': 0.0,
            'Average_Hold_Time': 0,
            'Win_Rate': 0.0,
            'Profit_Factor': 0.0,
            'Average_Win': 0.0,
            'Average_Loss': 0.0,
            'Risk_Reward_Ratio': 0.0
        }
```

```python
def plot_results(results_a, results_b):
    """Create visualizations for trading results"""
    plt.style.use('default')  # Use default style instead of seaborn
    fig = plt.figure(figsize=(15, 10))

    # Plot 1: Returns Distribution
    plt.subplot(2, 2, 1)
    plt.hist([r['Final_Return'] for r in results_a], alpha=0.5, label='Strategy A', bins=20, color='blue')
    plt.hist([r['Final_Return'] for r in results_b], alpha=0.5, label='Strategy B', bins=20, color='red')
    plt.title('Distribution of Returns', fontsize=12, pad=10)
    plt.xlabel('Return', fontsize=10)
    plt.ylabel('Frequency', fontsize=10)
    plt.legend()
    plt.grid(True, alpha=0.3)

    # Plot 2: Win Rates
    plt.subplot(2, 2, 2)
    win_rates_a = [r.get('Win_Rate', 0) for r in results_a]
    win_rates_b = [r.get('Win_Rate', 0) for r in results_b]

    bp = plt.boxplot([win_rates_a, win_rates_b], labels=['Strategy A', 'Strategy B'], patch_artist=True)
    plt.title('Win Rate Distribution', fontsize=12, pad=10)
    plt.ylabel('Win Rate', fontsize=10)
    plt.grid(True, alpha=0.3)

    # Color the boxplots
    bp['boxes'][0].set_facecolor('blue')
    bp['boxes'][0].set_alpha(0.6)
    bp['boxes'][1].set_facecolor('red')
    bp['boxes'][1].set_alpha(0.6)

    # Plot 3: Risk-Reward Ratio
    plt.subplot(2, 2, 3)
    rr_ratio_a = [r.get('Risk_Reward_Ratio', 0) for r in results_a]
    rr_ratio_b = [r.get('Risk_Reward_Ratio', 0) for r in results_b]

    bp2 = plt.boxplot([rr_ratio_a, rr_ratio_b], labels=['Strategy A', 'Strategy B'], patch_artist=True)
    plt.title('Risk-Reward Ratio Distribution', fontsize=12, pad=10)
    plt.ylabel('Risk-Reward Ratio', fontsize=10)
    plt.grid(True, alpha=0.3)

    bp2['boxes'][0].set_facecolor('blue')
    bp2['boxes'][0].set_alpha(0.6)
    bp2['boxes'][1].set_facecolor('red')
    bp2['boxes'][1].set_alpha(0.6)

    # Plot 4: Trade Counts
    plt.subplot(2, 2, 4)
    plt.hist([r.get('Total_Trades', 0) for r in results_a], alpha=0.5, label='Strategy A', bins=20, color='blue')
    plt.hist([r.get('Total_Trades', 0) for r in results_b], alpha=0.5, label='Strategy B', bins=20, color='red')
    plt.title('Distribution of Trade Counts', fontsize=12, pad=10)
    plt.xlabel('Number of Trades', fontsize=10)
    plt.ylabel('Frequency', fontsize=10)
    plt.legend()
    plt.grid(True, alpha=0.3)

    plt.tight_layout(pad=2.0)
    plt.show()

def print_detailed_results(results_a, results_b):
    """Print detailed analysis of trading results"""
    print("\n=== Strategy A Detailed Results ===")
    df_a = pd.DataFrame([{**r, 'Strategy': 'A'} for r in results_a])
    print(f"Number of Stocks Traded: {len(df_a)}")
    print(f"Average Return: {df_a['Final_Return'].mean():.2%}")
    print(f"Return Std Dev: {df_a['Final_Return'].std():.2%}")
    print(f"Total Trades: {df_a['Total_Trades'].sum()}")
    print(f"Average Trades per Stock: {df_a['Total_Trades'].mean():.1f}")
    print(f"Overall Win Rate: {df_a['Win_Rate'].mean():.2%}")
    print(f"Average Risk-Reward Ratio: {df_a['Risk_Reward_Ratio'].mean():.2f}")
    print(f"Average Profit Factor: {df_a['Profit_Factor'].mean():.2f}")
    print(f"Best Trade: {df_a['Best_Trade'].max():.2%}")
    print(f"Worst Trade: {df_a['Worst_Trade'].min():.2%}")
    print(f"Average Hold Time: {df_a['Average_Hold_Time'].mean():.1f} days")

    print("\n=== Strategy B Detailed Results ===")
    df_b = pd.DataFrame([{**r, 'Strategy': 'B'} for r in results_b])
    print(f"Number of Stocks Traded: {len(df_b)}")
    print(f"Average Return: {df_b['Final_Return'].mean():.2%}")
    print(f"Return Std Dev: {df_b['Final_Return'].std():.2%}")
    print(f"Total Trades: {df_b['Total_Trades'].sum()}")
    print(f"Average Trades per Stock: {df_b['Total_Trades'].mean():.1f}")
    print(f"Overall Win Rate: {df_b['Win_Rate'].mean():.2%}")
    print(f"Average Risk-Reward Ratio: {df_b['Risk_Reward_Ratio'].mean():.2f}")
    print(f"Average Profit Factor: {df_b['Profit_Factor'].mean():.2f}")
    print(f"Best Trade: {df_b['Best_Trade'].max():.2%}")
    print(f"Worst Trade: {df_b['Worst_Trade'].min():.2%}")
    print(f"Average Hold Time: {df_b['Average_Hold_Time'].mean():.1f} days")

    # Save detailed results to CSV
    pd.concat([df_a, df_b]).to_csv('all_results.csv', index=False)
```

```python
import matplotlib.pyplot as plt

# Trading simulation
start_date = "2024-10-01"
end_date = "2024-10-25"

results_a = []
results_b = []

for ticker in top_100['Ticker']:
    print(f"Trading simulation for {ticker}...")
    result_a = simulate_trading(ticker, start_date, end_date, 'A')
    result_b = simulate_trading(ticker, start_date, end_date, 'B')

    if result_a:
        result_a_analysis = analyze_results(pd.DataFrame(result_a['Trades']))
        if result_a_analysis:
            results_a.append({
                'Ticker': ticker,
                **result_a_analysis,
                'Final_Return': result_a['Final_Return']
            })

    if result_b:
        result_b_analysis = analyze_results(pd.DataFrame(result_b['Trades']))
        if result_b_analysis:
            results_b.append({
                'Ticker': ticker,
                **result_b_analysis,
                'Final_Return': result_b['Final_Return']
            })

# Print and plot results
print_detailed_results(results_a, results_b)
plot_results(results_a, results_b)
```
